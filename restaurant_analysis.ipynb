## **About the dataset**

The dataset contains information about 9,551 restaurants, with 21 columns capturing various attributes related to restaurant operations, location, and customer feedback. The key columns include:

* Restaurant ID: A unique identifier for each restaurant.

* Restaurant Name: The name of the restaurant.

* Country Code: Numeric representation of the country where the restaurant is located.

* City: The city where the restaurant is situated.

* Address: The restaurant's address.

* Locality & Locality Verbose: Details about the locality and its extended description.

* Longitude & Latitude: Geospatial coordinates of the restaurant.

* Cuisines: A list of cuisines offered by the restaurant.

* Average Cost for Two: Estimated cost for a meal for two people.

* Currency: The currency used for pricing.

* Has Table Booking: Whether the restaurant offers table booking options.

* Has Online Delivery: Indicates whether the restaurant provides online delivery services.

* Is Delivering Now: Shows whether the restaurant is currently delivering.

* Switch to Order Menu: A flag indicating availability of the "order menu."

* Price Range: Price range rating (typically a numeric scale).

* Aggregate Rating: The overall rating given by customers.

* Rating Color & Rating Text: Color and text associated with the restaurant's rating.

* Votes: The number of votes the restaurant's ratings have received.


The dataset offers a rich view into restaurant features across multiple countries and cities, providing insights into pricing, customer ratings, and service offerings. It is ideal for tasks such as customer preference analysis, rating prediction, and location-based insights.

## Task 1 : Predictive Modelling

```
# Importing libraries to build regression model.

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```
# Selecting features x and target variable y.
x=df.drop("Aggregate rating",axis=1)
y=df["Aggregate rating"]
```

```
# Splitted the data for training and testing.
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Shape of X_train:",X_train.shape)
print("Shape of X_test:",X_test.shape)
print("Shape of y_train:",y_train.shape)
print("Shape of y_test:",y_test.shape)
```

```
# Scaling the features
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```
# Create a dictionary with different regression models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regression": RandomForestRegressor(n_estimators=100, random_state=42),
    "Decision Tree Regression": DecisionTreeRegressor(random_state=42)
}

# Initialize an empty list to store results
results = []

# Loop through each model, train it, and evaluate performance
for model_name, model in models.items():
    model.fit(X_train_scaled, y_train)  # Train the model using the training data
    y_pred = model.predict(X_test_scaled)  # Predict target values for the test data

    # Calculate evaluation metrics
    mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
    mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
    r2 = r2_score(y_test, y_pred)  # R-squared score

    # Add the results to the list
    results.append([model_name, mse, mae, r2])

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results, columns=["Model", "Mean Squared Error", "Mean Absolute Error", "r2_score"])

# Sort the DataFrame by R-squared score in descending order
results_df = results_df.sort_values(by="r2_score", ascending=False)

# Show the final results
results_df


```

The evaluation results show that Linear Regression and Random Forest Regression perform similarly, with Linear Regression having the highest R² score of 0.986, indicating excellent accuracy. Random Forest follows closely with an R² of 0.985, while Decision Tree Regression has the lowest R² of 0.975 and higher error metrics. Overall, Linear Regression is the best-performing model, with Random Forest as a close second.

```
# Select top most model as the best model.
best_model = results_df.iloc[0]["Model"]
best_model
```

```
# Save the model
import joblib
joblib.dump(best_model,"best_model.pkl")
print("Best model is saved as best_model.pkl")
```
